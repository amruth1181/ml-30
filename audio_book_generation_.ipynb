```python
!pip install PyPDF2 pyttsx3
!sudo apt-get install espeak-ng

import PyPDF2
import pyttsx3

# Initialize the speaker
speaker = pyttsx3.init()

# Open your book
with open('/content/drive/MyDrive/books/mathsfordsa â€” Make It Stick - THE MANTHAN SCHOOL - Page 1 - 328 | Flip PDF Online | PubHTML5.pdf', 'rb') as book:
    pdf_reader = PyPDF2.PdfReader(book)

    # Loop through pages and read/save
    full_text = ""
    for page_num in range(len(pdf_reader.pages)):
        text = pdf_reader.pages[page_num].extract_text()
        full_text += text

    # Save to an MP3 file
    speaker.save_to_file(full_text, 'my_audiobook.mp3')
    speaker.runAndWait()
```
