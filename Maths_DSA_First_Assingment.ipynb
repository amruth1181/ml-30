```python
import numpy as np

v = np.array([[3],
              [2]])

A = np.array([[2, 1],
              [0, 2]])

B = np.array([[3, 0],
              [0, -1]])

C = np.array([[1/2, -np.sqrt(3)/2],
              [np.sqrt(3)/2, 1/2]])

Av = np.matmul(A,v)
Bv = np.matmul(B,v)
Cv = np.matmul(C,v)
ACv = np.matmul(A,Cv)

print("Av =\n", Av)
print("\nBv =\n", Bv)
print("\nCv =\n", Cv)
print("\nACv =\n", ACv)
```

```python
import numpy as np
t = np.array([[1],
              [-2]])

A = np.array([[2, 1],
              [0, 2]])

B = np.array([[3, 0],
              [0, -1]])

C = np.array([[1/2, -np.sqrt(3)/2],
              [np.sqrt(3)/2, 1/2]])

ACt = np.matmul(A,(np.matmul(C,t)))
CAt = np.matmul(C,(np.matmul(A,t)))

ABt = np.matmul(A,(np.matmul(B,t)))
BAt = np.matmul(B,(np.matmul(A,t)))

BCt = np.matmul(B,(np.matmul(C,t)))
CBt = np.matmul(C,(np.matmul(B,t)))

print("ACt =\n", ACt)
print("\nCAt =\n", CAt)

print("\nABt =\n", ABt)
print("\nBAt =\n", BAt)

print("\nBCt =\n", BCt)
print("\nCBt =\n", CBt)

print("\nIs ACt equal to CAt?", np.allclose(ACt, CAt))
print("Is ABt equal to BAt?", np.allclose(ABt, BAt))
print("Is BCt equal to CBt?", np.allclose(BCt, CBt))
```

```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([3,2])
t = np.array([1,-2])

A = np.array([[2,0],[1,2]])
B = np.array([[3,0],[0,-1]])
C = np.array([[1/2,3/2],[-3/2,1/2]])

Av = np.matmul(A,v)
Bv = np.matmul(B,v)
Cv = np.matmul(C,v)


plt.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,label='v')
plt.quiver(0,0,Av[0],Av[1],angles='xy',scale_units='xy',scale=1,label='Av')
plt.quiver(0,0,Bv[0],Bv[1],angles='xy',scale_units='xy',scale=1,label='Bv')
plt.quiver(0,0,Cv[0],Cv[1],angles='xy',scale_units='xy',scale=1,label='Cv')

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid()
plt.legend()
plt.show()
```
