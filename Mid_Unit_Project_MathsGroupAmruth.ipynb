### 1. Characteristic Equation<a href="#1.-Characteristic-Equation" class="anchor-link"></a>

compute det(S−λI)=0, first we find S−λI S−λI = \begin{bmatrix}
4-λ & 2 \\\\ 2 & 1-λ \end{bmatrix}

\$\$ \det(S - \lambda I) = (4 - \lambda)(1 - \lambda) - (2)(2) \$\$

\$\$ = 4 - 4\lambda - \lambda + \lambda^2 \$\$

\$\$ = \lambda^2 - 5\lambda + 4 - 4 \$\$

\$\$ = \lambda^2 - 5\lambda \$\$

\$\$ \text{Characteristic equation:} \quad \lambda^2 - 5\lambda = 0 \$\$

### 2. Finding Eignenvalues<a href="#2.-Finding-Eignenvalues" class="anchor-link"></a>

\$\$ \lambda^2 - 5\lambda = 0 \$\$

\$\$ \text{Factors are:} \quad \lambda(\lambda - 5) = 0 \$\$

\$\$ \text{Hence Eigenvalues are:} \$\$

\$\$ \lambda_1 = 5 \$\$

\$\$ \lambda_2 = 0 \$\$

\$\$ \text{Largest eigenvalue:} \quad \lambda_1 = 5 \$\$

### 3. Finding Eigenvectors<a href="#3.-Finding-Eigenvectors" class="anchor-link"></a>

\$\$ S w = \lambda_1 w \$\$

\$\$ (S - 5I)w = 0 \$\$

\$\$ S - 5I = \begin{bmatrix} 4-5 & 2 \\\\ 2 & 1-5 \end{bmatrix} =
\begin{bmatrix} -1 & 2 \\\\ 2 & -4 \end{bmatrix} \$\$

\$\$ \begin{bmatrix} -1 & 2 \\\\ 2 & -4 \end{bmatrix} \begin{bmatrix} x

\$\$ - x + 2y = 0 \$\$

\$\$ x = 2y \$\$

Solving these equations by assuming y = 1, we get x = 2

\$\$ \text{Eigenvector form:} \quad w = \begin{bmatrix} 2 \\\\ 1
\end{bmatrix} \$\$

\$\$ = \frac{5}{5 + 0} \$\$

\$\$ = \frac{5}{5} \$\$

\$\$ = 1 \$\$

\$\$ = 100\\% \$\$

```python
S = np.array([[4, 2],
              [2, 1]])

eigenvalues, eigenvectors = np.linalg.eig(S)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors (columns):\n", eigenvectors)
```

```python
lambda1 = eigenvalues[0]
v1 = eigenvectors[:, 0]

lhs = S @ v1
rhs = lambda1 * v1

print("S @ v1       :", lhs)
print("lambda1 * v1 :", rhs)
print("Are they equal?", np.allclose(lhs, rhs))
```

```python

sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues  = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

print("Sorted eigenvalues (descending):", sorted_eigenvalues)
print("Corresponding eigenvectors (columns):\n", sorted_eigenvectors)

variance_explained = sorted_eigenvalues / sorted_eigenvalues.sum()
print("\nVariance explained by each PC:", np.round(variance_explained * 100, 2), "%")
```

## **EXERCISE 3 - Building PCA from scratch:**
## Part A:<a href="#Part-A:" class="anchor-link"></a>

-   Standardization: PCA is based on variance and selects directions
    with maximum variance. If one feature has a large scale (1000) and
    another has a small scale (0.1), the large-scale feature will
    dominate the principal components, even if it is not more
    informative. Standardizing the data (mean 0, variance 1) ensures all
    features contribute equally.

-   The Objective: The PCA objective is:

\[ \max\_{w} ; w^T Sw \]

where (S) is the covariance matrix and (w) is a unit vector ((\|w\| =
1)) representing the projection direction.

-   Dimensionality Reduction: To reduce from 10 features to 2
    dimensions, the projection matrix (W) has shape **10 × 2**,
    containing the top 2 eigenvectors of (S). The reduced data is
    obtained as: \[ Z = X W \]

### Exercise 3: Advanced — Building PCA from Scratch
### Part B: Jupyter Notebook

```python
np.random.seed(42)
X = np.random.rand(100, 5)
print("Shape of X:", X.shape)
```

```python
def standardize(X):
    """Centre each feature to mean=0 and scale to std=1."""
    mean = X.mean(axis=0)
    std  = X.std(axis=0)
    return (X - mean) / std

X_std = standardize(X)
print("Mean of standardised X (should be ~0):",  np.round(X_std.mean(axis=0), 6))
print("Std  of standardised X (should be ~1):",  np.round(X_std.std(axis=0),  6))
```

```python
S_manual = np.cov(X_std, rowvar=False)
print("Covariance matrix shape:", S_manual.shape)
print("\nCovariance matrix:\n", np.round(S_manual, 4))
```

```python
eigenvalues_m, eigenvectors_m = np.linalg.eig(S_manual)

idx = np.argsort(eigenvalues_m)[::-1]
eigenvalues_m  = eigenvalues_m[idx]
eigenvectors_m = eigenvectors_m[:, idx]

print("Sorted eigenvalues:", np.round(eigenvalues_m, 4))

var_exp = eigenvalues_m / eigenvalues_m.sum()
print("Variance explained:", np.round(var_exp * 100, 2), "%")
```

```python
W = eigenvectors_m[:, :2]
print("Projection matrix W shape:", W.shape)

Z_manual = X_std @ W
print("Projected dataset Z shape:", Z_manual.shape)
```

```python
pca = PCA(n_components=2)
Z_sklearn = pca.fit_transform(X_std)

for i in range(2):
    match_direct = np.allclose(Z_manual[:, i],  Z_sklearn[:, i], atol=1e-6)
    match_flipped = np.allclose(Z_manual[:, i], -Z_sklearn[:, i], atol=1e-6)
    status = "✓ match" if (match_direct or match_flipped) else "✗ mismatch"
    print(f"PC{i+1}: {status}  (direct={match_direct}, sign-flipped={match_flipped})")

print("\nCorrelation between manual and sklearn PCs:")
for i in range(2):
    corr = np.abs(np.corrcoef(Z_manual[:, i], Z_sklearn[:, i])[0, 1])
    print(f"  PC{i+1}: |corr| = {corr:.6f}  (should be 1.0)")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

for ax, Z, title in zip(axes,
                         [Z_manual, Z_sklearn],
                         ['Manual PCA', 'Scikit-Learn PCA']):
    ax.scatter(Z[:, 0], Z[:, 1], alpha=0.6, edgecolors='k', linewidths=0.4)
    ax.set_xlabel('PC 1', fontsize=12)
    ax.set_ylabel('PC 2', fontsize=12)
    ax.set_title(title, fontsize=13)
    ax.grid(True, linestyle='--', alpha=0.5)

plt.suptitle('Comparison: Manual PCA vs Scikit-Learn PCA\n(shapes should be identical, signs may differ)',
             fontsize=12)
plt.tight_layout()
plt.savefig('exercise3_pca_comparison.png', dpi=120)
plt.show()

print("\nSklearn variance explained:",
      np.round(pca.explained_variance_ratio_ * 100, 2), "%")
```
